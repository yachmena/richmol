```python
from scipy import optimize, constants
import matplotlib.pyplot as plt


bohr_to_angstrom = constants.value('Bohr radius') * 1e10
hartree_to_invcm = constants.value('hartree-inverse meter relationship') * 1e-2

def poten(q):
    r1, r2, a = q.T
    return f_hot_wat.potv(r1/bohr_to_angstrom, r2/bohr_to_angstrom, a) * hartree_to_invcm

q0 = optimize.minimize(poten, [1.0, 1.0, np.pi/2]).x
print("equilibrium q0:", q0)


r1 = np.linspace(0.5, 3.0, 100)
q = np.array(np.broadcast_to(q0, (len(r1), len(q0))))
q[:, 0] = r1
v = poten(q)

print("r1 potential cut")
plt.ylim(-2e3, 5e4)
plt.plot(r1, poten(q))
plt.show()


a = np.linspace(40, 180, 100) * np.pi/180.0
q = np.array(np.broadcast_to(q0, (len(a), len(q0))))
q[:, 2] = a
v = poten(q)

print("alpha potential cut")
plt.plot(a, poten(q))
plt.show()
```
