# Vibrational states of a triatomic molecule using traditional variational approach

```python
import numpy as np
import potentials.f_hot_wat
import potentials.h2s_tyuterev
from scipy import constants, optimize
import functools
import itertools
import operator
from numpy.polynomial.hermite import hermgauss, hermval, hermder
from numpy.polynomial.legendre import leggauss, legval, legder
from numpy.polynomial.chebyshev import chebgauss, chebval, chebder
from numpy.polynomial.laguerre import laggauss, lagval, lagder
import matplotlib.pyplot as plt
import jax
from jax import numpy as jnp
from jax.config import config
config.update("jax_enable_x64", True)
```

## General molecule KEO

The general form of vibrational kinetic energy operator, expressed in terms of internal coordinates $q_1,q_2, ...,q_{3N-6}$, is

$T = -\frac{1}{2}\sum_{\lambda,\mu}^{3N-6}\frac{\partial}{\partial q_\lambda}G_{\lambda,\mu}(q_1,q_2,...,q_{3N-6})\frac{\partial}{\partial q_\mu} + U(q_1,q_2,...,q_{3N-6})$,

where $G$ is symmetric kinetic energy matrix and $U$ is scalar pseudo-potential.

Here we implement calculation of $G$ and $U$ in `Gmat(q)` and `pseudo(q)` that accept as argument a list of internal coordinate values, i.e., `q.shape = (3N-6,)`. Vectorized versions are implemented in `batch_Gmat(q)` and `batch_pseudo(q)` that accept as argument a matrix of internal coordinate values on grid of `no_points` points, i.e., `q.shape = (no_points, 3N-6)`.

```python
G_to_invcm = constants.value('Planck constant') * constants.value('Avogadro constant') \
           * 1e16 / (4.0 * np.pi**2 * constants.value('speed of light in vacuum')) * 1e5


eps = jnp.array([[[ int((i - j) * (j - k) * (k - i) * 0.5)
    for k in range(3) ] for j in range(3) ] for i in range(3) ], dtype=jnp.float64)


def com(method):
    @functools.wraps(method)
    def wrapper_cm(*args, **kwargs):
        xyz = method(*args, **kwargs)
        com = jnp.dot(molecule.masses, xyz) / jnp.sum(molecule.masses)
        return xyz - com[jnp.newaxis, :]
    return wrapper_cm


@jax.jit
def gmat(q):
    xyz_g = jax.jacfwd(cart)(q)
    tvib = xyz_g
    xyz = cart(q)
    natoms = xyz.shape[0]
    trot = jnp.transpose(jnp.dot(eps, xyz.T), (2, 0, 1))
    ttra = jnp.array([jnp.eye(3, dtype=jnp.float64) for _ in range(natoms)])
    tvec = jnp.concatenate((tvib, trot, ttra), axis=2)
    masses_sq = np.array([np.sqrt(molecule.masses[i]) for i in range(natoms)])
    tvec = tvec * masses_sq[:, jnp.newaxis, jnp.newaxis]
    tvec = jnp.reshape(tvec, (natoms*3, len(q)+6))
    return jnp.dot(tvec.T, tvec)


@jax.jit
def Gmat(q):
    return jnp.linalg.inv(gmat(q)) * G_to_invcm

batch_Gmat = jax.jit(jax.vmap(Gmat, in_axes=0))


@jax.jit
def dGmat(q):
    return jax.jacfwd(Gmat)(q)

batch_dGmat = jax.jit(jax.vmap(dGmat, in_axes=0))


@jax.jit
def Detgmat(q):
    nq = len(q)
    return jnp.linalg.det(gmat(q)[:nq+3, :nq+3])


@jax.jit
def dDetgmat(q):
    return jax.grad(Detgmat)(q)


@jax.jit
def hDetgmat(q):
    return jax.jacfwd(jax.jacrev(Detgmat))(q)


@jax.jit
def pseudo(q):
    nq = len(q)
    G = Gmat(q)[:nq, :nq]
    dG = dGmat(q)[:nq, :nq, :]
    dG = jnp.transpose(dG, (0, 2, 1))
    det = Detgmat(q)
    det2 = det * det
    ddet = dDetgmat(q)
    hdet = hDetgmat(q)
    pseudo1 = (jnp.dot(ddet, jnp.dot(G, ddet))) / det2
    pseudo2 = (jnp.sum(jnp.diag(jnp.dot(dG, ddet))) + jnp.sum(G * hdet)) / det
    return (-3 * pseudo1 + 4 * pseudo2) / 32.0

batch_pseudo = jax.jit(jax.vmap(pseudo, in_axes=0))
```

## Internal coordinates and potential for triatomic molecule

In order to compute elements of the $G$ matrix and pseudopotential $U$, we need to define our chosen internal coordinates $q_1,q_2,...,q_{3N-6}$ via Cartesian coordinates of atoms, with arbitrarily chosen orientation of the Cartesian frame and the origin placed at the centre of mass. For a triatomic molecule, we can choose two internuclear bond distances $r_1$ and $r_2$ and bond angle $\alpha$, or adjacent angle $\rho=\pi-\alpha$, as three (i.e, $3N-6=3*3-6=3$) internal cooridnates. In `cart(q)`, we define Cartesian coordinates of atoms as functions of three internal coordinates, where we chose the $x$ axis bisecting the bond angle $\alpha$, the $z$ axis coinciding with the linearity axis when $\alpha=180^\circ$, and $y$ axis being always perpendicular to the plane of the molecule. `@com` decorator function shifts Cartesian coordinates to the centre of mass.

We use potential energy surface for water from `f_hot_wat` and H$_2$S from `h2s_tyuterev`, see 'hot_wat.ipynb' and 'h2s_tyuterev.ipynb' for details.

```python
bohr_to_angstrom = constants.value('Bohr radius') * 1e10
hartree_to_invcm = constants.value('hartree-inverse meter relationship') * 1e-2

h2o = type('H2O', (object,), {'masses' : np.array([15.990526, 1.007276, 1.007276])})
h2s = type('H2S', (object,), {'masses' : np.array([31.97207070, 1.00782505, 1.00782505])})


@com
def cart(q):
    r1, r2, rho = q
    a = jnp.pi - rho
    xyz = jnp.array([[0.0, 0.0, 0.0],
                    [r1 * jnp.cos(a/2), 0.0, r1 * jnp.sin(a/2)],
                    [r2 * jnp.cos(a/2), 0.0, -r2 * jnp.sin(a/2)]],
                    dtype=jnp.float64)
    return xyz


def poten(q):
    r1, r2, rho = q.T
    a = np.pi - rho
    if molecule.__name__ == 'H2O':
        return potentials.f_hot_wat.potv(r1/bohr_to_angstrom, r2/bohr_to_angstrom, a) * hartree_to_invcm
    elif molecule.__name__ == 'H2S':
        return potentials.h2s_tyuterev.h2s_tyuterev(r1, r2, a)
    else:
        raise ValueError(f"no potential found for molecule '{molecule.__name__}'")
```

Initialize molecule

```python
molecule = h2o
#molecule = h2s
```

## Analytic *modified* KEO for triatomic molecule

The pseudopotential $U$, as computed by `pseudo` or `batch_pseudo` functions, becomes singular when triatomic molecule approaches linear configuration, i.e., $U\sim \frac{1}{\sin^2\rho}$ at $\rho\rightarrow 0$. It can be shown analytically that the cancellation of singularity occurs when the wave function approaches linearity as $\Psi\sim \sqrt{\sin\rho}$. Therefore, vibrational wave functions for quasi-linear triatomic molecules (i.e., molecules that can attain linear configuration) must be modelled as $\Psi_i(r_1,r_2,\rho)=\sqrt{\sin\rho}\phi_i(r_1,r_2,\rho)$. By multiplying the original KEO $T$ from the left by $\Psi_i$ and from the right by $\Psi_j$ we can derive an expression for a *modified* KEO $\tilde{T}$ such that its matrix elements in the basis of $\phi$ functions are equivalent to the matrix elements of the original KEO $T$ in the basis of $\Psi$ functions, i.e., $\langle\phi_i|\tilde{T}|\phi_j\rangle = \langle\Psi_i|T|\Psi_j\rangle$. The resulting expression is

$\tilde{T}=\sum_{\lambda,\mu}\frac{\partial}{\partial q_\lambda}\tilde{G}_{\lambda,\mu}^{(1)}\frac{\partial}{\partial q_\mu}+\sum_{\lambda}\frac{\partial}{\partial q_\lambda}\tilde{G}_{\lambda}^{(2)}+ \sum_{\lambda}\tilde{G}_{\lambda}^{(2)}\frac{\partial}{\partial q_\lambda}+\tilde{U}$.

The *modified* kinetic energy matrices $\tilde{G}^{(1)}$ and $\tilde{G}^{(2)}$ and pseudopotential $\tilde{U}$ are implemented in `Gmat1_sinrho(q)`, `Gmat2_sinrho(q)`, and `pseudo_sinrho(q)` functions.

```python
def Gmat1_sinrho(q):
    r1, r2, rho = q.T
    mx = molecule.masses[0]
    my = molecule.masses[1]
    g11 = np.sin(rho) * (1.0/(2*mx) + 1.0/(2*my))
    g12 = -np.sin(rho) * np.cos(rho) / (2.0*mx)
    g13 = np.sin(rho)**2 / (2.0*r2*mx)
    g22 = np.sin(rho) * (1.0/(2.0*mx) + 1.0/(2.0*my))
    g23 = np.sin(rho)**2 / (2.0*r1*mx)
    g33 = np.sin(rho) * (1.0/(2.0*r1**2*mx) + 1.0/(2.0*r2**2*mx) + 1.0/(2.0*r1**2*my) + 1.0/(2.0*r2**2*my) \
                         + np.cos(rho)/(r1*r2*mx))
    gmat = np.array([[g11, g12, g13], [g12, g22, g23], [g13, g23, g33]])
    return gmat * G_to_invcm


def Gmat2_sinrho(q):
    r1, r2, rho = q.T
    mx = molecule.masses[0]
    my = molecule.masses[1]
    g1 = np.sin(rho) * np.cos(rho) / (4*r2*mx)
    g2 = np.sin(rho) * np.cos(rho) / (4*r1*mx)
    g3 = np.cos(rho) * (1.0/(4.0*r1**2*mx) + 1.0/(4.0*r2**2*mx) + 1.0/(4.0*r1**2*my) + 1.0/(4.0*r2**2*my) \
                        + np.cos(rho)/(2.0*r1*r2*mx))
    gmat = np.array([g1, g2, g3])
    return gmat * G_to_invcm


def pseudo_sinrho(q):
    r1, r2, rho = q.T
    mx = molecule.masses[0]
    my = molecule.masses[1]
    u = -np.sin(rho) * (1.0/(4.0*r1**2*mx) + 1.0/(4.0*r2**2*mx) + 1.0/(4.0*r1**2*my) + 1.0/(4.0*r2**2*my))
    return u * G_to_invcm
```

## Quadratures and primitive basis functions

```python
def hermite_quadrature(npoints, q_ind, q0, h=0.001):
    x, w = hermgauss(npoints)

    G = Gmat(q0)

    # 7-point finite-difference rule for 2-order derivative of potential
    fdn_step = np.array([3*h, 2*h, h, 0.0, -h, -2*h, -3*h], dtype=np.float64)
    fdn_coef = np.array([2.0, -27.0, 270.0, -490.0, 270.0, -27.0, 2.0], dtype=np.float64)
    fdn_denom = 180.0

    # compute scaling
    q = np.array(np.broadcast_to(q0, (len(fdn_step), len(q0))))
    q[:, q_ind] = [q0[q_ind] + step for step in fdn_step]
    freq = np.dot(poten(q), fdn_coef) / (fdn_denom * h**2)
    scaling = np.sqrt(np.sqrt( 2.0 * np.abs(freq) / np.abs(G[q_ind, q_ind]) ))

    q = x / scaling + q0[q_ind]
    return q, w, scaling


def legendre_cos_quadrature(npoints):
    x, w = leggauss(npoints)
    q = np.arccos(x)
    return q, w, 1.0


def legendre_quadrature(npoints, a, b):
    x, w = leggauss(npoints)
    q = 0.5 * (b - a) * x + 0.5 * (a + b)
    return q, w, 1.0


def chebyshev1_quadrature(npoints, a, b):
    x, w = chebgauss(npoints)
    q = 0.5 * (b - a) * x + 0.5 * (a + b)
    return q, w, 1.0


def laguerre_quadrature(npoints, q_ind, q0, scaling=None, h=0.001):
    x, w = laggauss(npoints)

    if scaling is None:
        G = Gmat(q0)

        # 7-point finite-difference rule for 2-order derivative of potential
        fdn_step = np.array([3*h, 2*h, h, 0.0, -h, -2*h, -3*h], dtype=np.float64)
        fdn_coef = np.array([2.0, -27.0, 270.0, -490.0, 270.0, -27.0, 2.0], dtype=np.float64)
        fdn_denom = 180.0

        # compute scaling
        q = np.array(np.broadcast_to(q0, (len(fdn_step), len(q0))))
        q[:, q_ind] = [q0[q_ind] + step for step in fdn_step]
        freq = np.dot(poten(q), fdn_coef) / (fdn_denom * h**2)
        alpha = np.sqrt(2.0 * np.abs(freq) / np.abs(G[q_ind, q_ind]))
    else:
        alpha = scaling

    #q = np.sqrt(x / alpha + q0[q_ind])
    q = np.sqrt(x / alpha)
    return q, w, alpha


def hermite_functions(nmax, r0, scaling, q):
    x = (q - r0) * scaling
    sqsqpi = np.sqrt(np.sqrt(np.pi))
    c = np.diag([1.0 / np.sqrt(2.0**n * np.math.factorial(n)) / sqsqpi for n in range(nmax+1)])
    f = hermval(x, c)
    df = hermval(x, hermder(c, m=1))
    ddf = hermval(x, hermder(c, m=2))
    f_out = f
    df_out = (df - f * x) * scaling
    ddf_out = ((x**2 - 1) * f - 2 * x * df + ddf) * scaling**2
    return f_out, df_out, ddf_out


def legendre_cos_functions(nmax, q):
    x = np.cos(q)
    c = np.diag([np.sqrt((2.0 * n + 1)/2.0) for n in range(nmax+1)])
    f = legval(x, c)
    df = legval(x, legder(c, m=1))
    ddf = legval(x, legder(c, m=2))
    f_out = f
    df_out = df * (-np.sin(q))
    ddf_out = ddf * np.sin(q)**2 - df * np.cos(q)
    return f_out, df_out, ddf_out


def legendre_functions(nmax, a, b, q):
    x = 2 * q / (b - a) - (a + b) / (b - a)
    c = np.diag([np.sqrt(0.5 * (2.0 * n + 1)) for n in range(nmax+1)])
    f = legval(x, c)
    df = legval(x, legder(c, m=1)) * 2 / (b - a)
    return f, df


def chebyshev1_functions(nmax, a, b, q):
    x = 2 * q / (b - a) - (a + b) / (b - a)
    norm0 = np.sqrt(1.0 / np.pi)
    norm = np.sqrt(2.0 / np.pi)
    c = np.diag([ norm0 if n==0 else norm for n in range(nmax+1)])
    f = chebval(x, c)
    df = (chebval(x, chebder(c, m=1)) - 0.25 / (1.0 - x*x) * f) * 2 / (b - a)
    return f, df


def laguerre_functions(nmax, r0, scaling, q):
    x = scaling * q**2
    c = np.diag([np.sqrt(2 * scaling) for n in range(nmax+1)])
    f = lagval(x, c)
    df = lagval(x, lagder(c, m=1))
    ddf = lagval(x, lagder(c, m=2))
    f_out = f
    df_out = (df * 2 - f) * q * scaling
    ddf_out = ((x - 1) * f + (2 - 4 * x) * df + 4 * x * ddf) * scaling
    return f_out, df_out, ddf_out
```

## 1D solutions

```python
ncoo = 3
npoints = [100, 100, 100]
nmax = [30, 30, 30]
emax = 100000.0

vmin = optimize.minimize(poten, [1.0, 1.0, np.pi/2])
q0 = vmin.x
v0 = vmin.fun
print("equilibrium q0:", q0, v0)

psi1d = []
quad1d = []

for icoo in range(ncoo):

    # quadrature grid and basis functions

    if icoo in (0, 1): # stretching
        q, w, scaling = hermite_quadrature(npoints[icoo], icoo, q0)
        fpsi = functools.partial(hermite_functions, nmax[icoo], q0[icoo], scaling)
        psi, dpsi, ddpsi = fpsi(q)
    else: # bending
        q, w, scaling = legendre_cos_quadrature(npoints[icoo])
        fpsi = functools.partial(legendre_cos_functions, nmax[icoo])
        psi, dpsi, ddpsi = fpsi(q)

    psi *= np.sqrt(w)
    dpsi *= np.sqrt(w)
    ddpsi *= np.sqrt(w)

    print(f"quadrature range for icoo {icoo}: {np.min(q)} ... {np.max(q)}, scaling {scaling}")

    # operators on grid

    coords = np.array(np.broadcast_to(q0, (len(q), len(q0))))
    coords[:, icoo] = q

    if icoo in (0, 1): # stretching
        V = poten(coords)
        G = batch_Gmat(coords)
        U = batch_pseudo(coords)
        vme = np.einsum('ig,g,jg->ij', psi, V, psi, optimize='optimal', dtype=np.float64)
        gme = 0.5 * np.einsum('ig,g,jg->ij', dpsi, G[:, icoo, icoo], dpsi, optimize='optimal', dtype=np.float64)
        ume = np.einsum('ig,g,jg->ij', psi, U, psi, optimize='optimal', dtype=np.float64)
        hme = gme + vme + ume
        smat = np.einsum('ig,jg->ij', psi, psi, optimize='optimal', dtype=np.float64)
        si = np.linalg.inv(smat)
        hme = np.dot(si, hme)
    else: # bending
        sinrho = np.sin(q)
        V = poten(coords)
        G1 = Gmat1_sinrho(coords)
        G2 = Gmat2_sinrho(coords)
        U = pseudo_sinrho(coords)
        vme = np.einsum('ig,g,jg->ij', psi, V * sinrho, psi, optimize='optimal', dtype=np.float64)
        gme1 = np.einsum('ig,g,jg->ij', dpsi, G1[icoo, icoo], dpsi, optimize='optimal', dtype=np.float64)
        gme2 = np.einsum('ig,g,jg->ij', dpsi, G2[icoo], psi, optimize='optimal', dtype=np.float64) \
             + np.einsum('ig,g,jg->ij', psi, G2[icoo], dpsi, optimize='optimal', dtype=np.float64)
        ume = np.einsum('ig,g,jg->ij', psi, U, psi, optimize='optimal', dtype=np.float64)
        hme = gme1 + gme2 + vme + ume
        smat = np.einsum('ig,jg,g->ij', psi, psi, sinrho, optimize='optimal', dtype=np.float64)
        si = np.linalg.inv(smat)
        hme = np.dot(si, hme)
        psi /= np.sqrt(sinrho)
        dpsi /= np.sqrt(sinrho)
        ddpsi /= np.sqrt(sinrho)

    # plot operators

    print(f"potential for icoo {icoo}")
    plt.ylim(-2e3, 5e4)
    plt.plot(q, V)
    plt.show()
    if icoo == 2: # bending
        print(f"G1-matrix for icoo {icoo}")
        plt.plot(q, G1[icoo, icoo])
        plt.show()
        print(f"G2-matrix for icoo {icoo}")
        plt.plot(q, G2[icoo])
        plt.show()
    else: # stretching
        print(f"G-matrix for icoo {icoo}")
        plt.plot(q, G[:, icoo, icoo])
        plt.show()
    print(f"pseudo-potential for icoo {icoo}")
    plt.plot(q, U)
    plt.show()

    # energies and wavefunctions

    enr, vec = np.linalg.eigh(hme)

    no_states = len([e for e in enr if e-enr[0]<=emax])

    vec = vec[:, :no_states]
    psi1d.append( np.transpose(np.dot(vec.T, (psi, dpsi, ddpsi)), (1, 0, 2)) )
    quad1d.append((q, w))

    # print energies and plot wave functions

    print(f"zero-energy for icoo {icoo}:", enr[0])
    print(f"energies for icoo {icoo}:\n", enr[:no_states] - enr[0])

    print(f"wave functions for icoo {icoo}")
    for f, i in zip(psi,range(4)):
        plt.plot(q, f)
    plt.show()

    print(f"wave function derivatives for icoo {icoo}")
    for f, i in zip(dpsi,range(4)):
        plt.plot(q, f)
    plt.show()

    print(f"wave function 2nd derivatives for icoo {icoo}")
    for f, i in zip(ddpsi,range(4)):
        plt.plot(q, f)
    plt.show()
```

## 3D quadrature

Generate direct product quadrature from 1D quadratures of each coordinate.
Evaluate operators and basis functions on product quadrature grid

```python
# direct product of 1D quadratures

#  map indices of points in 1D quadratures to points' indices in 3D quadrature
ind = [[i for i in range(len(q))] for q, w in quad1d]
ind = np.array([elem for elem in itertools.product(*ind)]).T

# generate 3D quadrature points
q = np.array([q[i] for (q, w), i in zip(quad1d, ind)]).T

print(f"number of quadrature points: {len(q)}")

# evaluate potential on quadrature grid

V = poten(q)

# truncate quadrature to include points with (V - min(V)) <= Vtol
Vmin = np.min(V)
Vmax = np.max(V)
#Vtol = 1e50
Vtol = 1e5
ind2 = np.where(V - Vmin <= Vtol)
print(f"min and max of potential: {Vmin, Vmax}")
print(f"number of quadrature points for V-Vmin <= {Vtol}: {len(ind2[0])}")
q = q[ind2]
V = V[ind2]
ind = ind.T[ind2].T

# evaluate KEO on quadrature grid

sinrho = np.sin(q[:, 2])
G1_sinrho = Gmat1_sinrho(q)
G2_sinrho = Gmat2_sinrho(q)
U_sinrho = pseudo_sinrho(q)

G = batch_Gmat(q)
U = batch_pseudo(q)

# evaluate basis functions on quadrature grid

psi3d = [ psi.T[i].T for psi, i in zip(psi1d, ind)]

print("psi3d shape:", [elem.shape for elem in psi3d])  # ((psi, psi', psi''), #functions, #points)
```

## 3D solutions using 1D solutions as basis

```python
coo = (0, 1, 2)
w = (2, 2, 1)
nmax = 10

psi = [elem[0] if icoo in coo else elem[0][:1,:] for icoo, elem in enumerate(psi3d)]
dpsi = [elem[1] if icoo in coo else elem[1][:1,:] for icoo, elem in enumerate(psi3d)]

n = [[i for i in range(len(f))] for f in psi]
ind = np.array([elem for elem in itertools.product(*n)
                if sum(n_ * w_ for n_, w_ in zip(elem, w)) <= nmax])

dpsi3 = []
for icoo, df in enumerate(dpsi):
    flist = [df if jcoo == icoo else f for jcoo, f in enumerate(psi)]
    prod = np.prod(np.array([f[i] for f, i in zip(flist, ind.T)]), axis=0)
    dpsi3.append(prod)

psi3 = np.prod(np.array([f[i] for f, i in zip(psi, ind.T)]), axis=0)

print(f"number of basis functions for icoo {coo}: {psi3.shape}")

# matrix elements

vme = np.einsum('ig,g,jg->ij', psi3, V * sinrho, psi3, optimize='optimal', dtype=np.float64)   
gme1 = np.einsum('lig,lmg,mjg->ij', dpsi3, G1_sinrho, dpsi3, optimize='optimal', dtype=np.float64)
gme2 = np.einsum('lig,lg,jg->ij', dpsi3, G2_sinrho, psi3, optimize='optimal', dtype=np.float64) \
     + np.einsum('ig,lg,ljg->ij', psi3, G2_sinrho, dpsi3, optimize='optimal', dtype=np.float64)
ume = np.einsum('ig,g,jg->ij', psi3, U_sinrho, psi3, optimize='optimal', dtype=np.float64)
smat = np.einsum('ig,jg,g->ij', psi3, psi3, sinrho, optimize='optimal', dtype=np.float64)

hme = gme1 + gme2 + vme + ume
si = np.linalg.inv(smat)
hme = np.dot(si, hme)

# energies and coefficients

enr, vec = np.linalg.eigh(hme)

no_states = len([e for e in enr if e-enr[0]<=emax])
vec = vec[:, :no_states]

print(f"zero-energy for icoo {coo}:", enr[0])
print(f"energies for icoo {coo}:\n", enr[:no_states] - enr[0])
```
